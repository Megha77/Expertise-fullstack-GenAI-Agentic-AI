# Copy List

```python
mylist = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
mylist1=mylist
mylist1
```

```python
print(id(mylist))
```

```python
print(id(mylist1))
```

```python
mylist2=mylist.copy()
mylist2
```

```python
print(id(mylist2)) # The memory address of mylist2 will be 
                   #different from mylist because mylist2 is a new list object (a shallow copy) created in a different memory location.
```

```python
mylist[0]=1
mylist
```

```python
mylist1  # mylist1 will be also impacted as it is pointing to the same list 
```

```python
mylist2 # Copy of list won't be impacted due to changes made on the original list 
```

# All / Any

```python
L1 = [1,2,3,4,0]
all(L1)  # Will Return false as one value is false (Value 0) 
```

```python
any(L1) # Will Return True as we have items in the list with True value 
```

```python
L2 = [1,2,3,4,True,False] 
all(L2)
```

```python
any(L2)
```

```python
L3 = [1,2,3,True]
all(L3) # Will return True as all items in the list are True
```

```python
any(L3)
```
