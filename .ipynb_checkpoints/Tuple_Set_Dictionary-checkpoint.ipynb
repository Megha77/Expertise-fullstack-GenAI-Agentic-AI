# Add & Remove Items 

```python
myset
```

```python
myset.add('NINE')
myset
```

```python
myset.update(['TEN','ELEVEN','TWELVE']) # Add multiple item to a set using update
myset
```

```python
myset.remove('NINE') # raise error if element not present and we removing
myset
```

```python
myset.discard('TEN') # Does not raise error if element not present and we removing
myset
```

```python
myset.clear()
myset
```

```python
del myset
myset
```

# Copy Set

```python
myset = {'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight'} 
myset 
```

```python
myset1=myset
myset1
```

```python
id(myset), id(myset1)
```

```python
my_set=myset.copy()
myset
```

```python
id(my_set)
```

```python
myset.add('nine')
myset
```

```python
myset1 # myset1 will be also impacted as it is pointing to the same Set 
```

```python
my_set # Copy of the set won't be impacted due to changes made on the original set
```
