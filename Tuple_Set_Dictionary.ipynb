# Set Operation

```python
#Union
```

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {8,9,10}
```

```python
A|B # Union
```

```python
A.union(B)
```

```python
A.union(B,C)
```

```python
'''
Update set calling update() method with union of A,B,C
'''
A.update(B,C)
A
```

```python
# Intersection
```

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A & B
```

```python
A.intersection(B)
```

```python
'''
Update set calling intersection_update() with intersection of B
'''
A.intersection_update(B)
A
```

```python
# Diffrence
```

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A-B # set of elements that are only in A but not in B
```

```python
A.difference(B)
```

```python
B-A # set of elements that are only in B but not in A
```

```python
B.difference(A)
```

```python
"""
Updates the set calling the difference_update() method with the difference of se
For below example Set B will be updated with the difference of B & A.
"""
B.difference_update(A)
B
```

```python
# Symmetric Difference
```

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A ^ B # Symmetric difference (Set of elements in A and B but not in both. "EXCLUDE Common element"
```

```python
A.symmetric_difference(B)
```

```python
"""
Updates the set calling the symmetric_difference_update() method with the symmet
For below example Set A will be updated with the symmetric difference of A & B.
"""
A.symmetric_difference_update(B)
A
```

```python
# Subset , Superset & Disjoint
```

```python
A = {1,2,3,4,5,6,7,8,9}
B = {3,4,5,6,7,8}
C = {10,20,30,40}
```

```python
B.issubset(A)  # Set B is said to be the subset of set A if all elements of B are A
```

```python
A.issuperset(B)  # Set A is said to be the superset of set B if all elements of B are in A
```

```python
C.isdisjoint(A) # Two sets are said to be disjoint sets if they have no common element
```

```python
B.isdisjoint(A) # Two sets are said to be disjoint sets if they have no common e
```
